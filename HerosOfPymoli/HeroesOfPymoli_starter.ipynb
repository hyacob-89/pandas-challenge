```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
```

## Player Count

* Display the total number of players


```python
# Count unique SN data to get number of players
list_of_players = purchase_data['SN'].nunique()
total_players = [{"Total Players": list_of_players}]
total_players_df = pd.DataFrame(total_players)
total_players_df
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
# Count unique Item ID data to get number of unique items
number_of_unique_items = purchase_data['Item ID'].nunique()

# Calculate average purchase price
list_of_price = purchase_data['Price']
average_purchase_price = round(list_of_price.mean(),2)

# Calculate total number of purchases
total_number_of_purchases = purchase_data['Purchase ID'].count()

# calculate total revenue
total_revenue = round(list_of_price.sum(),2)

# Bring calculations into a dicitonary list
purchasing_analysis_dict = [{"Number of Unique Items": number_of_unique_items,
"Average Purchase Price": f"${average_purchase_price}",
"Total Number of Purchases": total_number_of_purchases,
"Total Revenue": f"${total_revenue:,}"}]

# Converrt dict list into a dataframe
purchasing_analysis_df = pd.DataFrame(purchasing_analysis_dict)
purchasing_analysis_df
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
# Group data by gender
gender_groupby = purchase_data.groupby('Gender')
demographics_count = gender_groupby[['SN']].nunique().rename(columns={"SN": "Total Count"})
percent_of_players = demographics_count['Total Count'] / list_of_players
demographics_count['Percentage of Players'] = percent_of_players.map("{:.2%}".format)
demographics_count
```

## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# Calculate purchase count and total purchase value
purchase_analysis_df = purchase_data.groupby('Gender')
purchase_count = purchase_analysis_df[['SN']].count().rename(columns={'SN': 'Purchase Count'})
purchase_value = purchase_analysis_df[['Price']].sum().rename(columns={'Price': 'Total Purchase Value'})

# Calculate Avg Total Purchase per Person
purchase_per_person_df = purchase_data.groupby(['SN', 'Gender']).sum()
avg_purchase_per_person = purchase_per_person_df[['Price']].rename(columns={'Price': 'Avg Total Purchase per Person'})
total_purchase_per_person = avg_purchase_per_person.groupby('Gender').mean()

# Merge the Dataframes on Gender
merge_tables_v1 = pd.merge(purchase_count, purchase_value, on='Gender').sort_values('Purchase Count', ascending=False)
merge_tables_v1['Average Purchase Price'] = merge_tables_v1['Total Purchase Value']/merge_tables_v1['Purchase Count']
merge_tables_final = pd.merge(merge_tables_v1, total_purchase_per_person, on='Gender')

# Format values to float and, if necessary, currency
merge_tables_final['Total Purchase Value'] = merge_tables_final['Total Purchase Value'].map("${:.2f}".format)
merge_tables_final['Average Purchase Price'] = merge_tables_final['Average Purchase Price'].map("${:.2f}".format)
merge_tables_final['Avg Total Purchase per Person'] = merge_tables_final['Avg Total Purchase per Person'].map("${:.2f}".format)

merge_tables_final
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python


# Create names and bins for the data categorization process
group_names = ['<10', '10-14', '15-19', '20-24', '25-29', '30-34', '35-39', '40+']
age_bin = [0, 9, 14, 19, 24, 29, 34, 39, 100]

# Using the bins, i created a new column titled 'Age Ranges'
purchase_data['Age Ranges'] = pd.cut(purchase_data["Age"], age_bin, labels=group_names)

# Groupped the main dataframe by age ranges and retrieved the total count
age_demo_raw_df = purchase_data.groupby('Age Ranges')
age_count_df = age_demo_raw_df[['SN']].nunique().rename(columns={"SN": "Total Count"})

# Brought in 'list of players' data frame to help calculate the percentages
age_count_df['Total Players'] = list_of_players

# Calculate the percentage of players by age range
percentage_of_players_age = (age_count_df['Total Count']/age_count_df['Total Players'])
age_count_df['Percentage of Players'] = percentage_of_players_age.map("{:.2%}".format)

age_demo_clean = age_count_df[['Total Count','Percentage of Players']]
age_demo_clean
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#Calculate purchase count and total purchase value
purchasing_analysis_raw_df = purchase_data.groupby('Age Ranges')
purchasing_analysis_count_df = purchasing_analysis_raw_df[['SN']].count().rename(columns={'SN': 'Purchase Count'})
purchasing_analysis_value_df = purchasing_analysis_raw_df[['Price']].sum().rename(columns={'Price': 'Total Purchase Value'})
purchasing_analysis_count_df['Total Players'] = list_of_players

# Calculate Avg Total Purchase per Person
pa_per_person_group = purchase_data.groupby(['Age Ranges','SN'])
pa_per_person_rename = pa_per_person_group[['Price']].sum().rename(columns={'Price': 'Avg Total Purchase per Person'})
purchasing_analysis_total_purchase_per_person = pa_per_person_rename.groupby('Age Ranges').mean()

# Merge the data frames on Age Ranges
df_merge_1 = pd.merge(purchasing_analysis_count_df, purchasing_analysis_value_df, on='Age Ranges')
df_merge_1["Average Purchase Price"] = df_merge_1['Total Purchase Value'] / df_merge_1['Purchase Count']
df_merge_final = pd.merge(df_merge_1, purchasing_analysis_total_purchase_per_person, on='Age Ranges')

# Format values to float and, if necessary, currency.
df_merge_final['Total Purchase Value'] = df_merge_final['Total Purchase Value'].map("${:.2f}".format)
df_merge_final['Average Purchase Price'] = df_merge_final['Average Purchase Price'].map("${:.2f}".format)
df_merge_final['Avg Total Purchase per Person'] = df_merge_final['Avg Total Purchase per Person'].map("${:.2f}".format)
                                                                 
# Select the columns for the dataframe
purchasing_analysis_final = df_merge_final[['Purchase Count', 'Average Purchase Price', 'Total Purchase Value', 'Avg Total Purchase per Person']]

purchasing_analysis_final                            
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Calculate purchase count and total purchase value
top_spenders_df = purchase_data.groupby('SN')
top_spenders_count_df = top_spenders_df[['SN']].count().rename(columns={"SN": "Purchase Count"})
top_spenders_value_df = top_spenders_df[['Price']].sum().rename(columns={"Price": "Total Purchase Value"})
top_spenders_count_df['Total Players'] = list_of_players

# Merge the data frames on SN
top_spenders_merge_1 = pd.merge(top_spenders_count_df, top_spenders_value_df, on='SN')
top_spenders_merge_1["Average Purchase Price"] = top_spenders_merge_1['Total Purchase Value'] / top_spenders_merge_1['Purchase Count']
top_spenders_merge_final = top_spenders_merge_1.sort_values('Total Purchase Value', ascending=False)

# Format values to float and, if necessary, currency.
top_spenders_merge_final['Total Purchase Value'] = top_spenders_merge_final['Total Purchase Value'].map("${:.2f}".format)
top_spenders_merge_final['Average Purchase Price'] = top_spenders_merge_final['Average Purchase Price'].map("${:.2f}".format)

# Select the columns for the dataframe
top_spenders_analysis_final = top_spenders_merge_final[['Purchase Count','Average Purchase Price', 'Total Purchase Value']]
top_spenders_analysis_final.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Calculate purchase count and total purchase value
MPI_df = purchase_data.groupby(['Item ID', 'Item Name','Price'])
MPI_count_df = MPI_df[['SN']].count().rename(columns={"SN": "Purchase Count"}).reset_index(drop=False)
MPI_value_df = MPI_df[['Price']].sum().rename(columns={"Price": "Total Purchase Value"})

# Merge the data frames on Item ID
MPI_merge_1 = pd.merge(MPI_count_df, MPI_value_df, on=['Item ID'])
MPI_merge_final = MPI_merge_1.sort_values("Purchase Count", ascending=False)

# Format values to float and, if necessary, currency.
MPI_merge_final['Total Purchase Value'] = MPI_merge_final['Total Purchase Value'].map("${:.2f}".format)
#MPI_merge_final['Price'] = MPI_merge_final['Price'].map('${:,.2f}'.format)

# Select the columns for the dataframe
MPI_analysis_final = MPI_merge_final[['Item ID', 'Item Name', 'Purchase Count', 'Price', 'Total Purchase Value']]
MPI_analysis_final.set_index(['Item ID', 'Item Name']).head()
```
